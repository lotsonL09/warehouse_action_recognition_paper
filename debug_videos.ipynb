```python
from pathlib import Path
from decord import VideoReader,cpu
from sklearn.model_selection import train_test_split
```

```python
root=Path.cwd()
videos_dir=root/"datasets"/"mini_dataset"
original_video_dir=root/"datasets"/"trimmed_videos"
```

```python
def debug_dir(videos_dir:Path):

    videos_path=videos_dir.glob("*/*/*.mp4")

    for video in videos_path:
        vr=VideoReader(str(video))
        frames = len(vr)
        if frames != 150:
            print(f"Video path: {video} | Length: {frames}")
```

```python
debug_dir(original_video_dir)
```

```python
def rename_videos(videos_dir:Path):
    for person in videos_dir.iterdir():
        for activity in person.iterdir():
            i=1
            for video in activity.iterdir():
                new_name=activity/f"video_{i}.mp4"
                video.rename(new_name)
                i+=1
```

```python
rename_videos(original_video_dir)
```

```python
import shutil
```

```python
def split_data_to_train_test(videos_dir:Path,dst_dir:Path):

    dst_train=dst_dir/"train"
    dst_train.mkdir(parents=True,exist_ok=True)
    
    dst_test=dst_dir/"test"
    dst_test.mkdir(parents=True,exist_ok=True)

    for person in videos_dir.iterdir():
        for activity in person.iterdir():
            # dst_activity_dir=dst_dir/activity.name
            # dst_activity_dir.mkdir(parents=True,exist_ok=True)
            videos=list(activity.iterdir())
            
            train_data,test_data=train_test_split(videos,test_size=0.2)
            for video in train_data:
                dst_activity_train=dst_train/activity.name
                dst_activity_train.mkdir(parents=True,exist_ok=True)
                dst_name=dst_activity_train/f"{person.name}_{video.name}"
                shutil.copy(video,dst_name)
            
            for video in test_data:
                dst_activity_test=dst_test/activity.name
                dst_activity_test.mkdir(parents=True,exist_ok=True)
                dst_name=dst_activity_test/f"{person.name}_{video.name}"
                shutil.copy(video,dst_name)


            
```
